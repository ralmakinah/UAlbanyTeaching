<h3>Expert Systems</h3>
<p>Rule based computer program that tried to capture and use human expertise on tasks.
    <p>They could work well - e.g. IBM's Deep Blue system that in 1996 defeated the world champion Gary Kasparov in chess.
        
<p> They had significant shortcomings:
<ul>
<li>They had to be created by hand. Every case, every scenario, every contingency had to be manually coded into them.</li>
    <li>This made them narrow in scope, and often brittle.</li>
    <li> They did not scale well.</li>
</ul>
    

<h3>Neural Networks and Deep Learning</h3>
---> Another AI Winter but !!! 

In the 1980s, several researchers revived and revised an <em>artificial neural networks</em> (ANN) style of AI.
<p>One that was centered on <em> machine learning</em> (ML).
    <p> Weaknesses of early ANN models:
        <ul>
            <li>Lack of computing power</li>
            <li>Inability to learn and/or represent certain classes of problems.</li>
            </ul>

    

By the mid 1980s, solutions were forthcoming for both of these problems.
<p>With the advent of <b>seminconductors</b> (basically computers and related electronics on chips) <em>Moore's Law</em> was observed. That is, every 18 months the number of transistors that could be put on a computer (memory, etc.) chip doubled.
    <p>This results in increased speed and computing power for chip based computers.
        This took care of the first problem (and still is...)
<p> in 1986, David Rumlhart, Geoffrey Hinton, and Ronald Williams published their work on <b> error backpropagation</b>. This was a generalization of the way that the Perceptron worked. But, it could deal with much more complex classes of problems.
<p>Within a few years, backpropagation-based machine learning systems had become the dominant model in AI. And have remained so ever since.

<p>
        Other trend and discoveries from around the same time have led us to the modern main trends in machine learning.
        <ul>
            <li>Kunihiko discovered <b>pattern recognition</b> techniques that became the basis for <b>convolutional neural networks</b> (CNN).</li>
            <li>John Hopfield laid the basis for <b>recurrent neural networks</b> - networks that can handle sequences of data (e.g. a series of images).</li>
            <li>Yann LeCun joined <b> convolutional neural networks with backpropagation</b>. This became the basis for modern <b>deep learning</b>.</li>
            <li>Later, LeCun made significant improvements into how the backpropogation algorithms work (added descent algorithms).</li>
            <li>Slightly different, but becoming increasingly important, Christopher Watkins, invented <b>Q-Learning</b>. This was an important foundation for <b>reinforcement learning</b>. In recent years, similar to LeCun's merger of backpropogation with CNNs to create deep learning, deep learning is merging with reinforcement learning to create <em>Deep Reinforcement Learning</em> to help solve additional types of problems.</li>
            </ul>
            
</p>

<h3>Factors Contributing to AI/ML today</h3>
In addition to the algorithms, several things are contributing to the popularity/utility of AI/ML today.
<ul>
    <li>Data - these machine learning models need data to learn from. The near universal use of computers (and data gaterhing sensors like cameras) provides this necessary data.</li>
    <li>Lots of fast computers, storage capacity, and high-speed networks. The infrastructure that holds this all together.</li>
    <li>GPUs - graphics processing units. These were originally were created to render computer graphics in video games. but, these same calculations, and the ability to do them <em>in parallel</em> make them also useful for AI/ML learning and calculations. (And for crypto mining, but that's a different story...)</li>
